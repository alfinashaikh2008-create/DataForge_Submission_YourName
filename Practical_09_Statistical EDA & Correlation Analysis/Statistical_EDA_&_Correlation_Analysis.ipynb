```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset_cleaned.xlsx.csv")
```

```python
df.head()
```

```python
numeric_df = df.select_dtypes(include=['number'])

print(numeric_df.head())
```

```python
correlation_matrix = numeric_df.corr()

print(correlation_matrix)
```

```python
correlation_matrix.to_csv("correlation_matrix.csv", index=True)

print("correlation_matrix.csv created successfully!")
```

```python
corr = correlation_matrix.abs()

# Remove self-correlations
np.fill_diagonal(corr.values, 0)

highest = corr.unstack().idxmax()
highest_value = corr.unstack().max()

print("Highest Positive Correlation:")
print("Variables:", highest)
print("Correlation:", highest_value)
```
