```python
import pandas as pd
```

```python
df = pd.read_csv("dataset_cleaned.xlsx.csv")
```

```python
print(df.isnull().sum())
```

```python
threshold = len(df) * 0.5
df = df.dropna(axis=1, thresh=threshold)

print(df.columns)
```

```python
numeric_columns = df.select_dtypes(include="number").columns

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

print(df.isnull().sum())
```

```python
region_sales = df.groupby("Region")["Sales"].mean().reset_index()

print(region_sales)
```

```python
region_sales.to_csv("region_sales_aggregated.csv", index=False)

print("region_sales_aggregated.csv created successfully!")
```
