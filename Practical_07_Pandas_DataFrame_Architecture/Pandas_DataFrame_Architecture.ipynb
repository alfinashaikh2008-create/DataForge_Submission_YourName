```python
with open("data_info.txt", "w") as f:
    f.write("=== DATASET INFORMATION ===\n\n")

    df.info(buf=f)

    f.write("\n\n=== DATASET SHAPE ===\n")
    f.write(str(df.shape))

    f.write("\n\n=== MISSING VALUES ===\n")
    f.write(str(df.isnull().sum()))

    f.write("\n\n=== MEMORY USAGE ===\n")
    f.write(str(df.memory_usage(deep=True)))
    f.write("\n\nTotal Memory Used: " + str(df.memory_usage(deep=True).sum()) + " bytes")
```
